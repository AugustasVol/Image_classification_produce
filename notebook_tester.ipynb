# A visual test of the model
### Using sample pictures

```python
import cnn_keras
import images_dir_labels
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
paths_X, paths_y = images_dir_labels.list_paths_labels("samples")
y = pd.get_dummies(paths_y)
```

```python
image_reader = images_dir_labels.image_iter(paths_X, y, batch_size=1 , scale=0.25, normalize=False)
```

```python
model = cnn_keras.cnn((256,256, 3), 6)
model.load_weights("model_weights_0.25_6")
```

```python
%%capture --no-stdout --no-display
for batch_X, batch_y in image_reader:
    plt.imshow(batch_X[0])
    num  = np.argmax(model.predict(batch_X))
    
    
    print(y.columns[num].split("/")[1])
    plt.show()
```
